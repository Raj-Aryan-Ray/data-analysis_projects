```python
# Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.



num = 3

if num <= 1:                        # the number is not grater than 1
    print("Given number has no positive divisors ")
else:
    is_prime = True
    
    for i in range(2, num):
        if num % i == 0:
            is_prime = False        # hear num%i == 0 means it is not a Prime 
            break
    
    print(is_prime)                # hear num%i != 0 means it is a Prime
    
```

```python
# Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.


import random

# Generate two random numbers
num1 = random.randint(1, 10)  # Random integer between 1 and 10
num2 = random.randint(1, 10)  # Random integer between 1 and 10


correct_product = num1 * num2 # product of two numbers
print(f"What is the product of {num1} and {num2}?")


try:
    user_answer = int(input("Your answer: "))# Get user input
    if user_answer == correct_product:  # Check if the answer is correct
        print("Correct Answer")
    else:
        print("Wrong Answer. The correct answer is ",correct_product)
except ValueError:
    print("Invalid input. Please enter a whole number")
```

```python
#Problem Statement (Question): Hospital Billing System
#A hospital charges its patients based on the following:
#•	Room charges per day: ₹2,000
#•	Doctor consultation fee (flat): ₹1,500
#•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
#•	Medicine charges: Total cost of medicines provided
#•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
#Write a Python program to calculate the final bill for a patient who:
#•	Stayed for 4 days
#•	Had 3 lab tests
#•	Had medicine charges of ₹2,400



# Given charges
ROOM_CHARGE_PER_DAY = 2000
DOCTOR_CONSULTATION_FEE = 1500
LAB_TEST_COST_PER_TEST = 300
DISCOUNT_THRESHOLD = 10000
DISCOUNT_PERCENTAGE = 0.10 # 10%
# Patient details
days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400
# Calculate individual bills
total_room_charges = days_stayed * ROOM_CHARGE_PER_DAY
total_lab_test_charges = num_lab_tests * LAB_TEST_COST_PER_TEST

total_bill_before_discount = (total_room_charges + DOCTOR_CONSULTATION_FEE + total_lab_test_charges + medicine_charges)  # Calculattal bill before discount

# Apply discount if applicable
final_bill = total_bill_before_discount
if total_bill_before_discount > DISCOUNT_THRESHOLD:
    discount_amount = total_bill_before_discount * DISCOUNT_PERCENTAGE
    final_bill = total_bill_before_discount - discount_amount
    print("Total bill before discount: ₹",total_bill_before_discount)
    print("Discount applied (10%): ₹",discount_amount)
else:
    print("Total bill: ₹",total_bill_before_discount)

print("Final bill for the patient: ₹",final_bill)
```
