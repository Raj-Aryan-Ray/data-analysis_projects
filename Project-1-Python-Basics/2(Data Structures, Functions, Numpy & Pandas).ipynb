```python
#6.	Given a 'marks' variable, print the grade:
#- 90 and above: 'A'
#- 75–89: 'B'
#- 50–74: 'C'
#- Below 50: 'Fail'


marks = 67
if marks < 0 or marks > 100:       # marks is not between 0-100 outside the range
    print("The marks are outside the range(0-100).")
elif marks >= 90 and marks<=100:    # 90 and above: 'A'
   print( "grade is:-",'A')
elif marks >= 75 and marks<=89:     # 75–89: 'B'
    print( "grade is:-",'B')
elif marks >= 50 and marks<=74:      # 50–74: 'C'
    print( "grade is:-",'C')
else:                                 # else  Below 50: 'Fail'
    print( "grade is:-",'Fail')
```

```python
#7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.


number = 10
if number > 0:
    if number % 2 == 0:         # number is grater than 0 if it is even-number means print even-number
        print("Even Number.")
    elif number == 0:           # If the number is zero
        print("Zero.")
    else:
       print("Odd Number.")     # number is grater than 0 if it is Odd-number means print Odd-number
else:                           # number is less than 0 printing negative number
    print( "Negative Number.")
    
```

```python
#11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
#- Use .info(), .describe()
#- Select 'Name' and 'City' columns
#- Drop 'City' column
#- Fill any missing values in 'Salary' column with the mean
#- Remove any duplicate rows



import pandas as pd      
import numpy as np



# -Create a DataFrame manually with 10 rows and four columns
data = {
    'Name': ['Alex', 'Bella', 'Chris', 'David', 'Eve', 'Alex', 'Fiona', 'George', 'Hannah', 'Ian'],
    'Age': [25, 30, 22, 45, 35, 25, 28, 33, 40, 27],
    'City': ['London', 'Paris', 'Berlin', 'Rome', 'Madrid', 'London', 'Vienna', 'London', 'Oslo', 'Zurich'],
    # Including two missing values (NaN) in 'Salary' for the imputation step
    'Salary': [50000, 60000, np.nan, 80000, 75000, 50000, 62000, np.nan, 90000, 70000]}

print(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
#- info(),.describe()
df.info()
```

```python
df.describe()
```

```python
#-Select 'Name' and 'City' columns
df[['Name', 'City']]
```

```python
#- Drop 'City' column
df.drop('City', axis=1)
```

```python
#- Fill any missing values in 'Salary' column with the mean
m=df['Salary'].mean()
print(m)
```

```python
df['Salary'].fillna(value=m, inplace =True)
df
```

```python
#- Remove any duplicate rows
df.drop_duplicates(inplace=True)
df
```
