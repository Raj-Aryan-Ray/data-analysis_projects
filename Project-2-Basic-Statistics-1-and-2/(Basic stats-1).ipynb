```python
#●	Objective: To compute and analyze basic statistical measures for numerical columns in the dataset.
#●	Steps:
	#Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
	#Identify numerical columns in the dataset.
	#Calculate the mean, median, mode, and standard deviation for these columns.
	#Provide a brief interpretation of these statistics.
```

```python
#Load the dataset into a data analysis tool or programming environment 
import pandas as pd
df= pd.read_csv("sales_data_with_discounts.csv") # importing the data 
df
```

```python
#	Identify numerical columns in the dataset.
numerical_cols = df.select_dtypes(include=['number']).columns
print(f"Numerical columns identified: {list(numerical_cols)}")
```

```python
#	Calculate the mean, median, mode, and standard deviation for these columns.
# Mean
print("MEAN:-",df['Total Sales Value'].mean())  #Represents the typical value. If the Mean is much higher or lower than the Median, it suggests the presence of outliers or skewness.
```

```python
# Median
print("MEDIAN:-",df['Total Sales Value'].median())  #Provides the true center of the dataset, as it's unaffected by extreme outliers.

```

```python
# Mode
print("MODE:-",df['Total Sales Value'].mode()) #  Identifies the most common observation. For continuous data, the mode may not be meaningful unless the data is highly clustered.

```

```python
# Standard Deviation
print("STANDARD DIVIATION:-",df['Total Sales Value'].std()) # Measures the risk or volatility. A low standard deviation means data points are clustered tightly around the mean; a high standard deviation means they are widely spread.
```

```python
# Variance
print("VAR:-",df['Total Sales Value'].var())
```

```python
# Group by
print(df.groupby('Total Sales Value')['Avg Price'].mean())
```
