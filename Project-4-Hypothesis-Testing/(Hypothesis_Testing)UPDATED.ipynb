```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuration for plots
sns.set(style='whitegrid')
np.random.seed(42)
print('Libraries imported and seed set.')
```

```python
# Fixed cell: Defining required variables and calculating p-value
from scipy import stats
import math

# Data from the problem
sample_mean = 3050
n = 25
alpha = 0.05
mu = 1000 + (5 * 600)
sigma = 5 * 25

# Calculate Z-statistic
z_statistic = (sample_mean - mu) / (sigma / math.sqrt(n))

# Calculating the p-value for a right-tailed test
p_val = 1 - stats.norm.cdf(z_statistic)

print(f'Z-statistic: {z_statistic}')
print(f'P-value: {p_val}')

if p_val < alpha:
    print(f'Since p-value ({p_val:.4f}) < alpha ({alpha}), we reject the Null Hypothesis.')
else:
    print(f'Since p-value ({p_val:.4f}) > alpha ({alpha}), we fail to reject the Null Hypothesis.')
```

```python
from scipy import stats

# Calculate the critical Z-value for a right-tailed test at alpha = 0.05
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)

print(f'Alpha level: {alpha}')
print(f'Critical Z-value (right-tailed): {z_critical:.4f}')
```

```python
# 1. State the Hypotheses
# Null Hypothesis (H0): The mean weekly operating cost is Rs. 4,000 (W = 1000 + 5*600)
# Alternative Hypothesis (H1): The mean weekly operating cost is greater than Rs. 4,000

# Given Data
sample_mean = 3050
n = 25
alpha = 0.05

# Theoretical Mean (mu) from the model W = 1000 + 5X
x_mean_units = 600
mu = 1000 + (5 * x_mean_units)

# Standard Deviation (sigma)
# The problem states sigma for X is 25, and cost W = 1000 + 5X.
# Therefore, the standard deviation of the cost is 5 * 25
sigma = 5 * 25

print(f'Theoretical Mean (mu): {mu}')
print(f'Sample Mean (x_bar): {sample_mean}')
print(f'Population Standard Deviation (sigma): {sigma}')
print(f'Sample Size (n): {n}')
```

```python
# 2. Calculate the Test Statistic (Z)
import math

z_statistic = (sample_mean - mu) / (sigma / math.sqrt(n))

print(f'Calculated Z-statistic: {z_statistic:.4f}')
```

```python
# 3. Determine the Critical Value and P-value
from scipy import stats

# Since the claim is that costs are higher, this is a right-tailed test
z_critical = stats.norm.ppf(1 - alpha)
p_value = 1 - stats.norm.cdf(z_statistic)

print(f'Critical Value (Z at alpha=0.05): {z_critical:.4f}')
print(f'P-value: {p_value:.4f}')
```

```python
# 4. Make a Decision
print(f'Decision Criteria: Reject H0 if Z-statistic > Z-critical or P-value < {alpha}')

if z_statistic > z_critical:
    decision = 'Reject the Null Hypothesis'
else:
    decision = 'Fail to Reject the Null Hypothesis'

print(f'Decision: {decision}')
```

```python
# 5. Conclusion
if z_statistic > z_critical:
    print('Conclusion: There is strong evidence to support the restaurant owners\' claim that the weekly operating costs are higher than the model suggests.')
else:
    print('Conclusion: There is not enough evidence to support the claim. The observed costs are not significantly higher than the theoretical model.')
```
