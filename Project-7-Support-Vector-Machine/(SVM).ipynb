```python
#data= df(includes="floate64","int64").columns
data=df.select_dtypes(include=["float64","int64"]).columns
```

```python
for col in data:
    
    # Calculate quartiles
    q1 = np.percentile(df[col], 25)
    q3 = np.percentile(df[col], 75)
    iqr = q3 - q1
    
    # Upper and lower bounds
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr   # ❗ corrected (was + in your code)
    
    # Cap outliers
    df.loc[df[col] > upper, col] = upper
    df.loc[df[col] < lower, col] = lower
    
    # Plot after outlier treatment
    plt.figure()
    plt.boxplot(df[col], vert=False)
    plt.xlabel(col)
    plt.show()
   
```

```python
import numpy as np
import matplotlib.pyplot as plt

for i in data:
    
    # Quartiles
    q1 = np.percentile(df[i], 25)
    q3 = np.percentile(df[i], 75)
    iqr = q3 - q1

    # Bounds
    uw = q3 + (1.5 * iqr)
    lw = q1 - (1.5 * iqr)

    # Cap outliers
    df[i][df[i] > uw] = uw
    df[i][df[i] < lw] = lw

    # Boxplot
    plt.figure("/n")
    plt.boxplot(df[i], vert=False)
    plt.xlabel(i)
    plt.title(f"Boxplot of {i}")
    plt.show()

    # Histogram
    plt.figure()
    plt.hist(df[i], bins=30)
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.title(f"Histogram of {i}")
    plt.show("/n")
```

```python
#test3
Y=df["Drug Response"]
Y.head()
```

```python
X=df.iloc[:,0:4]
X.head()
```

```python
#test 2
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear',C=1)
model.fit(x_train,y_train)
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score
score1 = accuracy_score(y_train,y_pred_train)
score2 = accuracy_score(y_test,y_pred_test)
#test 3
print("Training score:", np.round(score1, 2))
print("Test score:", np.round(score2, 2))

# Evaluation metrics
accuracy = accuracy_score(y_test, y_pred_test)
precision = precision_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
f1 = f1_score(y_test, y_pred_test)

print("Accuracy :", np.round(accuracy,2))
print("Precision:",  np.round(precision,2))
print("Recall   :",  np.round(recall,2))
print("F1-score :",  np.round(f1,2))
```

```python
#test 6
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score

kernels = ['linear', 'poly', 'rbf']
C_values = [0.1, 1, 10]

for k in kernels:
    for c in C_values:
        model = SVC(kernel=k, C=c)
        model.fit(x_train, y_train)

        y_pred_test = model.predict(x_test)

        print(
            f"Kernel={k}, C={c} | "
            f"Train Score={model.score(x_train, y_train):.2f} | "
            f"Test Score={model.score(x_test, y_test):.2f} | "
            
            f"Accuracy={accuracy_score(y_test, y_pred_test):.2f} | "
            f"F1-score={f1_score(y_test, y_pred_test):.2f}"
        )
```

```python
gamma_values = [0.01, 0.1, 1]

for g in gamma_values:
    model = SVC(kernel='rbf', C=1, gamma=g)
    model.fit(x_train, y_train)
    print(f"Gamma={g}, Test Accuracy={model.score(x_test, y_test):.2f}")
```

```python
# ==============================
# SVM Visualization for ALL X pairs
# ==============================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from itertools import combinations

# Target
y = Y

# List of feature names
feature_names = X.columns

# Loop through all feature pairs
for feat1, feat2 in combinations(feature_names, 2):

    # Select two features
    X_2d = X[[feat1, feat2]]

    # Train Linear SVM
    svm = SVC(kernel='rbf', C=0.1)
    svm.fit(X_2d, y)

    # Create mesh grid
    x_min, x_max = X_2d.iloc[:, 0].min() - 1, X_2d.iloc[:, 0].max() + 1
    y_min, y_max = X_2d.iloc[:, 1].min() - 1, X_2d.iloc[:, 1].max() + 1

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300),
        np.linspace(y_min, y_max, 300)
    )

    # Predict grid
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot
    plt.figure(figsize=(7, 5))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
    plt.scatter(
        X_2d.iloc[:, 0],
        X_2d.iloc[:, 1],
        c=y,
        cmap='viridis',
        edgecolors='k'
    )

    plt.xlabel(feat1)
    plt.ylabel(feat2)
    plt.title(f"SVM Decision Boundary: {feat1} vs {feat2}")
    plt.colorbar(label="Drug Response")
    plt.show()
```

```python
#task 7
SVM Strengths & Weaknesses (Based on EDA)

Strengths

Works well with overlapping and non-linear data

Good generalization (train and test scores are close)

RBF kernel captures patterns missed by linear models

Weaknesses

Sensitive to C 

Less interpretable

Computationally expensive for large datasets

Practical Implications of SVM

Suitable for medium-sized classification problems

Effective in text, medical, and pattern recognition tasks

Not ideal for very large or highly noisy datasets

Preferred when accuracy is more important than explainability
```
