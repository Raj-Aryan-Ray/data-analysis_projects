```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3)
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=46,random_state=42,max_features=0.7,max_depth=4,max_samples=24)
rf.fit(x_train,y_train)
y_train_pred=rf.predict(x_train)
y_test_pred=rf.predict(x_test)
training_score=accuracy_score(y_train,y_train_pred)
test_score=accuracy_score(y_test,y_test_pred)
pre=precision_score(y_test,y_test_pred,average='weighted')
recall=recall_score(y_test,y_test_pred,average='weighted')
f1=f1_score(y_test,y_test_pred,average='weighted')

print(f"training accuracy score:{training_score:.2f}")
print(f"test_score:{test_score:.2f}")
print(f"precall score:{pre:.2f}")
print(f"recall score:{recall:.2f}")
print(f"f1 score:{f1:.2f}")
```

```python
Bagging (Bootstrap Aggregating) is an ensemble method where multiple models are trained independently and in parallel on different random samples of the dataset. The final prediction is obtained by averaging or majority voting. Bagging mainly helps in reducing variance and overfitting.

Boosting is an ensemble method where models are trained sequentially, and each new model focuses more on the errors made by previous models. The final prediction is a weighted combination of all models. Boosting helps in reducing bias and improving accuracy.

Difference:
Bagging trains models independently to reduce variance, while Boosting trains models sequentially to reduce bias by focusing on misclassified samples.
```

```python
Handling Imbalanced Data

Imbalanced data occurs when one class has significantly fewer samples than others.
It can be handled using resampling techniques such as oversampling the minority class or undersampling the majority class.
Another method is using class weights, where higher importance is given to the minority class during model training. 
Additionally, appropriate evaluation metrics like precision, recall, F1-score, and ROC-AUC should be used instead of accuracy alone.
```
